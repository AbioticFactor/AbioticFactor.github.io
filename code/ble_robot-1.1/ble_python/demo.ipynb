```python
%load_ext autoreload
%autoreload 2

from ble import get_ble_controller
from base_ble import LOG
from cmd_types import CMD
import time
import numpy as np
import matplotlib.pyplot as plt
from scipy import pi
from scipy.fftpack import fft

LOG.propagate = False
```

```python
# Get ArtemisBLEController object
ble = get_ble_controller()

# Connect to the Artemis Device
ble.connect()
```

```python
time = []
pitch_a = []
roll_a = []
data = []
time_str = "T:"
pitch_str = "pitch:"
roll_str = "roll:"

def acc_5s_receiver(uiud, byte_arr):
    global time
    global pitch_a
    global roll_a
    global data
    
    new_data = ble.bytearray_to_string(byte_arr)
    print(new_data)
    
    # get time
    res = ""
    time_idx = new_data.index(time_str)
    for i in range(time_idx + len(time_str), len(new_data)):
        if new_data[i] == "|":
            break;
        res += new_data[i]
    print(res)
    time.append(int(res))
    
    # get pitch
    res = ""
    pitch_idx = new_data.index(pitch_str)
    for i in range(pitch_idx + len(pitch_str), len(new_data)):
        if new_data[i] == "|":
            break;
        res += new_data[i]
    pitch_a.append(float(res))
    
    # get roll
    res = ""
    roll_idx = new_data.index(roll_str)
    for i in range(roll_idx + len(roll_str), len(new_data)):
        if new_data[i] == "|":
            break;
        res += new_data[i]
    roll_a.append(float(res))
    
        
    
    data.append(new_data)
   

ble.start_notify(ble.uuid['RX_STRING'], acc_5s_receiver)
ble.send_command(CMD.GET_ACC_5s, "")
```

```python
print(len(time))
print(time[0:])
print(pitch_a[0:])
print(roll_a[0:])

ble.stop_notify(ble.uuid['RX_STRING'])
```

```python
sample_rate = len(time)/5
print(sample_rate)
N = (5 - 0) * sample_rate

#pitch 
plt.plot (time [0:], pitch_a [0:])
plt.xticks(np.arange(min(time), max(time), 1000))
plt.yticks(np.arange(min(pitch_a), max(pitch_a) + 5, 5))
plt.title ('Time Domain Signal for pitch')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()

frequency = np.linspace (0.0, 512, int (N/2))
freq_data = fft(pitch_a)
y = 2/N * np.abs(freq_data [0:int(N/2)])

plt.plot(frequency, y)
plt.title('Frequency domain Signal for pitch')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.show()

#roll
plt.plot (time [0:], roll_a [0:])
plt.xticks(np.arange(min(time), max(time), 1000))
plt.yticks(np.arange(min(roll_a), max(roll_a) + 5, 5))
plt.title ('Time Domain Signal for roll')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()

frequency = np.linspace (0.0, 512, int (N/2))
freq_data = fft(roll_a)
y = 2/N * np.abs(freq_data [0:int(N/2)])

plt.plot(frequency, y)
plt.title('Frequency domain Signal for roll')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.show()
```
